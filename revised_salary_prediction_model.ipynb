```python
# ============================================================
# INSTALL DEPENDENCIES (Run this cell first in Colab)
# ============================================================
!pip install catboost quantile-forest


# ============================================================
# SECTION 0: IMPORTS & SETUP
# ============================================================
import kagglehub
import numpy as np
import pandas as pd
import os
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from io import StringIO
from matplotlib.patches import Patch

from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
from quantile_forest import RandomForestQuantileRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from scipy.stats import randint as sp_randint

warnings.filterwarnings('ignore')
print(" All libraries loaded.")


# ============================================================
# SECTION 1: DATA LOADING (GitHub → Kaggle Fallback)
# ============================================================
# Source 1: GitHub mirror of aijobs.net - weekly updated, CC0,
#           no auth required, reliable raw CSV from Colab
# Source 2: Kaggle dataset - final fallback if GitHub is unreachable

GITHUB_URL = "https://raw.githubusercontent.com/foorilla/ai-jobs-net-salaries/main/salaries.csv"

df_raw = None
source = None

def normalize_aijobs_schema(df):
    """Normalize GitHub/aijobs.net columns into unified pipeline schema."""
    df = df.rename(columns={'salary_in_usd': 'salary_usd'})
    exp_map = {'EN': 1, 'MI': 3, 'SE': 7, 'EX': 12}
    df['years_experience'] = df['experience_level'].map(exp_map).fillna(3)
    df = df.dropna(subset=['salary_usd'])
    df = df[df['salary_usd'] > 0].reset_index(drop=True)
    return df

# --- Source 1: GitHub raw (primary) ---
print(" Fetching latest data from GitHub (foorilla/ai-jobs-net-salaries)...")
try:
    resp = requests.get(GITHUB_URL, timeout=15)
    resp.raise_for_status()
    df_raw = pd.read_csv(StringIO(resp.text))
    missing = [c for c in ['salary_in_usd', 'job_title', 'experience_level']
               if c not in df_raw.columns]
    if missing:
        raise ValueError(f"Unexpected schema - missing columns: {missing}")
    df_raw = normalize_aijobs_schema(df_raw)
    source = " LIVE - GitHub mirror of aijobs.net (weekly updated, CC0)"
    print(f" GitHub data loaded successfully: {df_raw.shape[0]:,} rows")

except Exception as e1:
    print(f" GitHub fetch failed: {e1}")
    print(" Falling back to Kaggle dataset...")

    try:
        path     = kagglehub.dataset_download(
                       'bismasajjad/global-ai-job-market-and-salary-trends-2025')
        csv_path = os.path.join(path, 'ai_job_dataset.csv')
        df_raw   = pd.read_csv(csv_path)
        if 'years_experience' not in df_raw.columns and 'experience_level' in df_raw.columns:
            exp_map = {'EN': 1, 'MI': 3, 'SE': 7, 'EX': 12}
            df_raw['years_experience'] = df_raw['experience_level'].map(exp_map).fillna(3)
        source = " FALLBACK - Kaggle (bismasajjad/global-ai-job-market-and-salary-trends-2025)"
        print(f" Kaggle fallback loaded: {df_raw.shape[0]:,} rows")

    except Exception as e2:
        raise RuntimeError(
            f"Both data sources failed.\n"
            f" GitHub error : {e1}\n"
            f" Kaggle error : {e2}"
        )

df = df_raw.copy()
print(f"\n Data source : {source}")
print(f" Shape        : {df.shape}")
print(f" Salary range : ${df['salary_usd'].min():,.0f} – ${df['salary_usd'].max():,.0f}")
print(f" Columns      : {list(df.columns)}")


# ============================================================
# SECTION 2: PREPROCESSING
# ============================================================
# Split data first, then fit encoder on training data only.
# This prevents test-set information from leaking into the encoder.

df_ml = df.copy()
cat_cols = df_ml.select_dtypes(include=['object']).columns.tolist()

# --- Step 1: Split raw (unencoded) data first ---
X_raw = df_ml.drop('salary_usd', axis=1)
y     = df_ml['salary_usd']

X_train_raw, X_test_raw, y_train, y_test = train_test_split(
    X_raw, y, test_size=0.2, random_state=42)

# --- Step 2: Fit encoder on TRAIN only, then apply to both ---
X_train = X_train_raw.copy()
X_test  = X_test_raw.copy()

for col in cat_cols:
    if col not in X_train.columns:
        continue
    le = LabelEncoder()
    X_train[col] = le.fit_transform(X_train[col].astype(str))
    # Map test categories; any unseen label gets encoded as -1
    le_classes = list(le.classes_)
    X_test[col] = X_test[col].astype(str).apply(
        lambda x: le_classes.index(x) if x in le_classes else -1)

# Also encode the full df for EDA visualisations only (not used in evaluation)
X_full = X_raw.copy()
for col in cat_cols:
    if col not in X_full.columns:
        continue
    le_full = LabelEncoder()
    X_full[col] = le_full.fit_transform(X_full[col].astype(str))

print(f" Preprocessing complete | Train: {X_train.shape[0]:,} | Test: {X_test.shape[0]:,}")
print(f" Features: {list(X_train.columns)}")


# ============================================================
# SECTION 3: EXPLORATORY DATA ANALYSIS
# ============================================================

# --- Top Highest Paid Jobs ---
highest_paid = df.groupby('job_title')['salary_usd'].max().sort_values(ascending=False).head(20)
plt.figure(figsize=(14, 6))
highest_paid.plot(kind='bar', color='skyblue')
plt.title('Top Highest Paid Job Titles (Actual Max)')
plt.ylabel('Salary (USD)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# --- Job Count by Employment Type ---
plt.figure(figsize=(10, 6))
emp_counts = df['employment_type'].value_counts()
ax = sns.barplot(x=emp_counts.index, y=emp_counts.values,
                 palette='viridis', edgecolor='black', alpha=0.8)
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 9),
                textcoords='offset points', fontsize=11, fontweight='bold')
plt.title('Global AI Job Count by Employment Type', fontsize=15, pad=20)
plt.xlabel('Employment Type', fontsize=12)
plt.ylabel('Number of Postings', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()

# --- Market Share by Company Size ---
plt.figure(figsize=(8, 8))
cs_counts = df['company_size'].value_counts()
explode   = [0.05 if i == 0 else 0 for i in range(len(cs_counts))]
plt.pie(cs_counts, labels=cs_counts.index, autopct='%1.1f%%', startangle=140,
        colors=sns.color_palette('pastel'), explode=explode, pctdistance=0.85,
        wedgeprops={'edgecolor': 'white', 'linewidth': 2})
plt.gcf().gca().add_artist(plt.Circle((0, 0), 0.70, fc='white'))
plt.title('Market Share by Company Size', fontsize=16, fontweight='bold')
plt.annotate(f'Total Jobs:\n{len(df)}', (0, 0), ha='center', va='center',
             fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()

# --- Salary distribution ---
plt.figure(figsize=(12, 5))
sns.histplot(df['salary_usd'], bins=60, kde=True, color='steelblue')
plt.title('Salary Distribution (USD)', fontsize=15)
plt.xlabel('Salary (USD)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

print("\n EDA Summary:")
print(f" Mean salary   : ${df['salary_usd'].mean():,.0f}")
print(f" Median salary : ${df['salary_usd'].median():,.0f}")
print(f" Std deviation : ${df['salary_usd'].std():,.0f}")
print(f" Min           : ${df['salary_usd'].min():,.0f}")
print(f" Max           : ${df['salary_usd'].max():,.0f}")
print(f" Skewness      : {df['salary_usd'].skew():.4f}")
print(f" Kurtosis      : {df['salary_usd'].kurt():.4f}")
q1 = df['salary_usd'].quantile(0.25)
q3 = df['salary_usd'].quantile(0.75)
print(f" Q1            : ${q1:,.0f}")
print(f" Q3            : ${q3:,.0f}")
print(f" IQR           : ${q3 - q1:,.0f}")


# ============================================================
# SECTION 4: TRAIN ALL MODELS (UNTUNED)
# ============================================================
print("\n" + "="*50)
print("  PHASE 1: UNTUNED MODEL TRAINING")
print("="*50)

exp_col = 'years_experience'

# --- Baseline: Linear Regression ---
print("\n Baseline: Linear Regression...")
baseline = LinearRegression()
baseline.fit(X_train, y_train)
base_pred_test = baseline.predict(X_test)
base_r2   = r2_score(y_test, base_pred_test)
base_mae  = mean_absolute_error(y_test, base_pred_test)
base_rmse = np.sqrt(mean_squared_error(y_test, base_pred_test))
base_ape  = (np.abs(y_test.values - base_pred_test) / y_test.values * 100).mean()
print(f" R²: {base_r2:.4f} | MAE: ${base_mae:,.0f} | RMSE: ${base_rmse:,.0f} | APE: {base_ape:.2f}%")

# --- XGBoost ---
print("\n XGBoost (untuned)...")
xgb = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=5,
                   random_state=42, verbosity=0)
xgb.fit(X_train, y_train)
xgb_pred_test = xgb.predict(X_test)
xgb_r2   = r2_score(y_test, xgb_pred_test)
xgb_mae  = mean_absolute_error(y_test, xgb_pred_test)
xgb_rmse = np.sqrt(mean_squared_error(y_test, xgb_pred_test))
xgb_ape  = (np.abs(y_test.values - xgb_pred_test) / y_test.values * 100).mean()
print(f" R²: {xgb_r2:.4f} | MAE: ${xgb_mae:,.0f} | RMSE: ${xgb_rmse:,.0f} | APE: {xgb_ape:.2f}%")

# --- LightGBM ---
print("\n LightGBM (untuned)...")
lgbm = LGBMRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
lgbm.fit(X_train, y_train)
lgbm_pred_test = lgbm.predict(X_test)
lgbm_r2   = r2_score(y_test, lgbm_pred_test)
lgbm_mae  = mean_absolute_error(y_test, lgbm_pred_test)
lgbm_rmse = np.sqrt(mean_squared_error(y_test, lgbm_pred_test))
lgbm_ape  = (np.abs(y_test.values - lgbm_pred_test) / y_test.values * 100).mean()
print(f" R²: {lgbm_r2:.4f} | MAE: ${lgbm_mae:,.0f} | RMSE: ${lgbm_rmse:,.0f} | APE: {lgbm_ape:.2f}%")

# --- CatBoost ---
print("\n CatBoost (untuned)...")
cat = CatBoostRegressor(iterations=300, learning_rate=0.05,
                        depth=6, random_seed=42, verbose=0)
cat.fit(X_train, y_train)
cat_pred_test = cat.predict(X_test)
cat_r2   = r2_score(y_test, cat_pred_test)
cat_mae  = mean_absolute_error(y_test, cat_pred_test)
cat_rmse = np.sqrt(mean_squared_error(y_test, cat_pred_test))
cat_ape  = (np.abs(y_test.values - cat_pred_test) / y_test.values * 100).mean()
print(f" R²: {cat_r2:.4f} | MAE: ${cat_mae:,.0f} | RMSE: ${cat_rmse:,.0f} | APE: {cat_ape:.2f}%")

# --- RF Quantile Regressor ---
print("\n RF Quantile Regressor (untuned)...")
rfq = RandomForestQuantileRegressor(n_estimators=200, max_depth=15,
                                    min_samples_leaf=5, random_state=42, n_jobs=-1)
rfq.fit(X_train, y_train)
rfq_pred_test = rfq.predict(X_test, quantiles=0.5)
rfq_lower     = rfq.predict(X_test, quantiles=0.1)
rfq_upper     = rfq.predict(X_test, quantiles=0.9)
rfq_coverage  = np.mean((y_test.values >= rfq_lower) & (y_test.values <= rfq_upper)) * 100
rfq_r2   = r2_score(y_test, rfq_pred_test)
rfq_mae  = mean_absolute_error(y_test, rfq_pred_test)
rfq_rmse = np.sqrt(mean_squared_error(y_test, rfq_pred_test))
rfq_ape  = (np.abs(y_test.values - rfq_pred_test) / y_test.values * 100).mean()
print(f" R²: {rfq_r2:.4f} | MAE: ${rfq_mae:,.0f} | RMSE: ${rfq_rmse:,.0f} | APE: {rfq_ape:.2f}%")
print(f" 80% Interval Coverage: {rfq_coverage:.1f}%")

print("\n All untuned models trained.")


# ============================================================
# SECTION 5: UNTUNED PERFORMANCE VISUALIZATIONS
# ============================================================

# --- R² Leaderboard ---
plt.figure(figsize=(13, 7))
u_labels = ['Baseline\n(Linear)', 'XGBoost', 'LightGBM', 'CatBoost', 'RF Quantile']
u_scores = [base_r2, xgb_r2, lgbm_r2, cat_r2, rfq_r2]
u_colors = ['royalblue', 'crimson', 'seagreen', 'darkorange', 'mediumpurple']
bars = plt.bar(u_labels, u_scores, color=u_colors, alpha=0.85, edgecolor='black')
for bar, score in zip(bars, u_scores):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.005,
             f'{score:.4f}', ha='center', fontweight='bold', fontsize=11)
best_i = u_scores.index(max(u_scores))
bars[best_i].set_edgecolor('gold')
bars[best_i].set_linewidth(3)
plt.annotate(' Best', xy=(best_i, u_scores[best_i] + 0.02),
             ha='center', fontsize=12, color='darkgoldenrod', fontweight='bold')
plt.ylim(0, max(u_scores) + 0.13)
plt.title('Phase 1 - Untuned Algorithm Leaderboard (R²)', fontsize=16)
plt.ylabel('R² Score (Higher is Better)')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()

# --- Salary Trend by Experience (using full dataset for visualisation only) ---
df['Baseline_Pred'] = baseline.predict(X_full)
df['XGB_Pred']      = xgb.predict(X_full)
df['LGBM_Pred']     = lgbm.predict(X_full)
df['CatBoost_Pred'] = cat.predict(X_full)
df['RFQ_Pred']      = rfq.predict(X_full, quantiles=0.5)

plt.figure(figsize=(14, 7))
exp_trend = df.groupby(exp_col).agg({
    'salary_usd':    'mean',
    'Baseline_Pred': 'mean',
    'XGB_Pred':      'mean',
    'LGBM_Pred':     'mean',
    'CatBoost_Pred': 'mean',
    'RFQ_Pred':      'mean',
}).sort_index()
plt.plot(exp_trend.index, exp_trend['salary_usd'],    label='Actual',      color='black',        marker='o', alpha=0.5, lw=2)
plt.plot(exp_trend.index, exp_trend['Baseline_Pred'], label='Baseline',    color='royalblue',    linestyle='--')
plt.plot(exp_trend.index, exp_trend['XGB_Pred'],      label='XGBoost',     color='crimson',      lw=2)
plt.plot(exp_trend.index, exp_trend['LGBM_Pred'],     label='LightGBM',    color='seagreen',     lw=2)
plt.plot(exp_trend.index, exp_trend['CatBoost_Pred'], label='CatBoost',    color='darkorange',   lw=2)
plt.plot(exp_trend.index, exp_trend['RFQ_Pred'],      label='RF Quantile', color='mediumpurple', lw=2, linestyle='-.')
plt.title('Untuned - Salary Trend by Experience: All Models', fontsize=15)
plt.xlabel('Years of Experience')
plt.ylabel('Average Salary (USD)')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# --- Distribution Fidelity ---
plt.figure(figsize=(15, 7))
sns.kdeplot(df['salary_usd'],    fill=True, color='gray',         label='Actual',      alpha=0.2)
sns.kdeplot(df['Baseline_Pred'], color='royalblue',  label='Baseline',    linestyle='--')
sns.kdeplot(df['XGB_Pred'],      color='crimson',    label='XGBoost',     lw=2)
sns.kdeplot(df['LGBM_Pred'],     color='seagreen',   label='LightGBM',    lw=2)
sns.kdeplot(df['CatBoost_Pred'], color='darkorange', label='CatBoost',    lw=2)
sns.kdeplot(df['RFQ_Pred'],      color='mediumpurple', label='RF Quantile', lw=2, linestyle='-.')
plt.title('Untuned - Distribution Fidelity: Model vs Real-World Salary Spread', fontsize=15)
plt.xlabel('Salary (USD)')
plt.legend()
plt.tight_layout()
plt.show()

# --- Multi-Market Precision Audit ---
top_locs = [l for l in df['company_location'].value_counts().nlargest(11).index if l != 'IL']
loc_stats = df[df['company_location'].isin(top_locs)].groupby('company_location').agg(
    Actual_Avg=('salary_usd',    'mean'),
    Actual_Std=('salary_usd',    'std'),
    Baseline_Avg=('Baseline_Pred','mean'),
    XGBoost_Avg=('XGB_Pred',     'mean'),
    LGBM_Avg=('LGBM_Pred',       'mean'),
    CatBoost_Avg=('CatBoost_Pred','mean'),
    RFQ_Avg=('RFQ_Pred',         'mean'),
).sort_values('Actual_Avg', ascending=False).head(10)

x, w = np.arange(len(loc_stats)), 0.14
fig, ax = plt.subplots(figsize=(20, 9))
ax.bar(x-2.5*w, loc_stats['Actual_Avg'],   w, label='Actual',      color='gray',        alpha=0.25, yerr=loc_stats['Actual_Std'], capsize=3)
ax.bar(x-1.5*w, loc_stats['Baseline_Avg'], w, label='Baseline',    color='royalblue',   alpha=0.8,  edgecolor='white')
ax.bar(x-0.5*w, loc_stats['XGBoost_Avg'],  w, label='XGBoost',     color='crimson',     alpha=0.8,  edgecolor='white')
ax.bar(x+0.5*w, loc_stats['LGBM_Avg'],     w, label='LightGBM',    color='seagreen',    alpha=0.8,  edgecolor='white')
ax.bar(x+1.5*w, loc_stats['CatBoost_Avg'], w, label='CatBoost',    color='darkorange',  alpha=0.8,  edgecolor='white')
ax.bar(x+2.5*w, loc_stats['RFQ_Avg'],      w, label='RF Quantile', color='mediumpurple',alpha=0.8,  edgecolor='white')
ax.set_title('Untuned - Multi-Model Salary Precision Audit by Market (Top 10)', fontsize=17, pad=15)
ax.set_ylabel('Salary (USD)')
ax.set_xticks(x)
ax.set_xticklabels(loc_stats.index, rotation=45, ha='right')
ax.set_ylim(0, loc_stats['Actual_Avg'].max() * 1.45)
ax.legend(loc='upper right', ncol=3, frameon=True, shadow=True)
ax.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()

# --- Feature Importance (XGBoost + CatBoost side by side) ---
fig, axes = plt.subplots(1, 2, figsize=(18, 7))
xgb_imp = pd.Series(xgb.feature_importances_, index=X_train.columns).sort_values(ascending=True)
xgb_imp.plot(kind='barh', ax=axes[0], color='crimson', edgecolor='black', alpha=0.8)
axes[0].set_title("XGBoost Feature Importance (Untuned)", fontsize=13)
axes[0].set_xlabel('Relative Importance')
cat_imp = pd.Series(cat.get_feature_importance(), index=X_train.columns).sort_values(ascending=True)
cat_imp.plot(kind='barh', ax=axes[1], color='darkorange', edgecolor='black', alpha=0.8)
axes[1].set_title("CatBoost Feature Importance (Untuned)", fontsize=13)
axes[1].set_xlabel('Relative Importance')
plt.suptitle('What Drives Salary Predictions? (Untuned)', fontsize=15, y=1.02)
plt.tight_layout()
plt.show()

# Print XGBoost importance scores for paper
print("\n XGBoost - Feature Importance Scores:")
for feat, score in xgb_imp.sort_values(ascending=False).items():
    print(f" {feat:<30} : {score:.4f}")

# --- LightGBM Feature Importance ---
lgbm_imp = pd.Series(lgbm.feature_importances_, index=X_train.columns).sort_values(ascending=True)
fig, ax = plt.subplots(figsize=(12, 7))
lgbm_imp.plot(kind='barh', ax=ax, color='seagreen', edgecolor='black', alpha=0.85)
top3 = lgbm_imp.nlargest(3).index
for patch, label in zip(ax.patches, lgbm_imp.index):
    if label in top3:
        patch.set_facecolor('darkgreen')
        patch.set_alpha(1.0)
for patch in ax.patches:
    width = patch.get_width()
    ax.text(width + lgbm_imp.max() * 0.01, patch.get_y() + patch.get_height() / 2,
            f'{int(width)}', va='center', ha='left', fontsize=9, fontweight='bold')
ax.set_title("LightGBM Feature Importance (Untuned)", fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Feature Importance Score (Split-based)', fontsize=12)
ax.set_ylabel('Feature', fontsize=12)
ax.grid(axis='x', linestyle='--', alpha=0.3)
ax.legend(handles=[
    Patch(facecolor='darkgreen', label='Top 3 drivers'),
    Patch(facecolor='seagreen', alpha=0.85, label='Other features')
], loc='lower right', frameon=True)
plt.tight_layout()
plt.show()

print("\n LightGBM - Top Feature Drivers:")
for rank, (feat, score) in enumerate(lgbm_imp.sort_values(ascending=False).head(5).items(), 1):
    print(f" {rank}. {feat:<30} score: {int(score)}")

# --- RF Quantile Prediction Intervals ---
print("\n Generating RF Quantile prediction interval plot...")
rfq_q10 = rfq.predict(X_full, quantiles=0.1)
rfq_q50 = rfq.predict(X_full, quantiles=0.5)
rfq_q90 = rfq.predict(X_full, quantiles=0.9)
sample  = df.nsmallest(300, exp_col)
s_idx   = sample.index
s_exp   = df.loc[s_idx, exp_col]
s_actual= df.loc[s_idx, 'salary_usd']
s_q50   = rfq_q50[df.index.get_indexer(s_idx)]
s_q10   = rfq_q10[df.index.get_indexer(s_idx)]
s_q90   = rfq_q90[df.index.get_indexer(s_idx)]

plt.figure(figsize=(14, 7))
plt.scatter(s_exp, s_actual, color='gray',        alpha=0.3, s=15, label='Actual Salary',        zorder=3)
plt.scatter(s_exp, s_q50,    color='mediumpurple', alpha=0.7, s=10, label='Predicted Median (Q50)', zorder=4)
for e in sorted(s_exp.unique()):
    mask = s_exp == e
    if mask.sum() > 0:
        plt.fill_between([e-0.3, e+0.3],
                         [np.percentile(s_q10[mask], 25)]*2,
                         [np.percentile(s_q90[mask], 75)]*2,
                         alpha=0.15, color='mediumpurple')
plt.title('RF Quantile - Salary Prediction with 80% Uncertainty Bands', fontsize=14)
plt.xlabel('Years of Experience')
plt.ylabel('Salary (USD)')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()


# ============================================================
# SECTION 6: UNTUNED SUMMARY
# ============================================================
print("\n" + "="*65)
print(f"{'PHASE 1 - UNTUNED PERFORMANCE SUMMARY':^65}")
print("="*65)
print(f"{'MODEL':<25} | {'R²':>6} | {'MAE':>10} | {'RMSE':>10} | {'APE':>8}")
print("-"*65)
for nm, r2, mae, rmse, ape in [
    ("Baseline (Linear)", base_r2,  base_mae,  base_rmse,  base_ape),
    ("XGBoost",           xgb_r2,   xgb_mae,   xgb_rmse,   xgb_ape),
    ("LightGBM",          lgbm_r2,  lgbm_mae,  lgbm_rmse,  lgbm_ape),
    ("CatBoost",          cat_r2,   cat_mae,   cat_rmse,   cat_ape),
    ("RF Quantile",       rfq_r2,   rfq_mae,   rfq_rmse,   rfq_ape),
]:
    print(f"{nm:<25} | {r2:>6.4f} | ${mae:>9,.0f} | ${rmse:>9,.0f} | {ape:>7.2f}%")
print("="*65)


# ============================================================
# SECTION 7: HYPERPARAMETER TUNING
# ============================================================
print("\n" + "="*50)
print("  PHASE 2: HYPERPARAMETER TUNING")
print("="*50)

# --- Tune LightGBM ---
print("\n Tuning LightGBM...")
lgbm_grid = GridSearchCV(
    LGBMRegressor(random_state=42),
    {'n_estimators':  [100, 200, 300],
     'learning_rate': [0.01, 0.05, 0.1],
     'max_depth':     [3, 5, 10],
     'num_leaves':    [20, 31, 50]},
    cv=3, scoring='r2', verbose=1, n_jobs=-1)
lgbm_grid.fit(X_train, y_train)
best_lgbm = lgbm_grid.best_estimator_
t_lgbm_pred   = best_lgbm.predict(X_test)
t_lgbm_r2   = r2_score(y_test, t_lgbm_pred)
t_lgbm_mae  = mean_absolute_error(y_test, t_lgbm_pred)
t_lgbm_rmse = np.sqrt(mean_squared_error(y_test, t_lgbm_pred))
t_lgbm_ape  = (np.abs(y_test.values - t_lgbm_pred) / y_test.values * 100).mean()
print(f" Best params : {lgbm_grid.best_params_}")
print(f" R²: {lgbm_r2:.4f} → {t_lgbm_r2:.4f}  ({((t_lgbm_r2-lgbm_r2)/lgbm_r2)*100:+.2f}%)")

# --- Tune XGBoost ---
print("\n Tuning XGBoost...")
xgb_grid = GridSearchCV(
    XGBRegressor(random_state=42, verbosity=0),
    {'n_estimators':     [100, 200, 300],
     'learning_rate':    [0.01, 0.05, 0.1],
     'max_depth':        [3, 5, 7],
     'subsample':        [0.7, 0.9, 1.0]},
    cv=3, scoring='r2', verbose=1, n_jobs=-1)
xgb_grid.fit(X_train, y_train)
best_xgb = xgb_grid.best_estimator_
t_xgb_pred   = best_xgb.predict(X_test)
t_xgb_r2   = r2_score(y_test, t_xgb_pred)
t_xgb_mae  = mean_absolute_error(y_test, t_xgb_pred)
t_xgb_rmse = np.sqrt(mean_squared_error(y_test, t_xgb_pred))
t_xgb_ape  = (np.abs(y_test.values - t_xgb_pred) / y_test.values * 100).mean()
print(f" Best params : {xgb_grid.best_params_}")
print(f" R²: {xgb_r2:.4f} → {t_xgb_r2:.4f}  ({((t_xgb_r2-xgb_r2)/xgb_r2)*100:+.2f}%)")

# --- Tune CatBoost ---
print("\n Tuning CatBoost...")
cat_grid = GridSearchCV(
    CatBoostRegressor(random_seed=42, verbose=0),
    {'iterations':    [200, 300, 500],
     'learning_rate': [0.01, 0.05, 0.1],
     'depth':         [4, 6, 8],
     'l2_leaf_reg':   [1, 3, 5]},
    cv=3, scoring='r2', verbose=1, n_jobs=-1)
cat_grid.fit(X_train, y_train)
best_cat = cat_grid.best_estimator_
t_cat_pred   = best_cat.predict(X_test)
t_cat_r2   = r2_score(y_test, t_cat_pred)
t_cat_mae  = mean_absolute_error(y_test, t_cat_pred)
t_cat_rmse = np.sqrt(mean_squared_error(y_test, t_cat_pred))
t_cat_ape  = (np.abs(y_test.values - t_cat_pred) / y_test.values * 100).mean()
print(f" Best params : {cat_grid.best_params_}")
print(f" R²: {cat_r2:.4f} → {t_cat_r2:.4f}  ({((t_cat_r2-cat_r2)/cat_r2)*100:+.2f}%)")

# --- Tune RF Quantile ---
print("\n Tuning RF Quantile (RandomizedSearchCV, n_iter=20)...")
rfq_random = RandomizedSearchCV(
    RandomForestQuantileRegressor(random_state=42, n_jobs=-1),
    {'n_estimators':      sp_randint(100, 400),
     'max_depth':         [10, 15, 20, None],
     'min_samples_leaf': sp_randint(2, 10),
     'max_features':      ['sqrt', 'log2', 0.7, 0.9]},
    n_iter=20, cv=3, scoring='r2', verbose=1, n_jobs=-1, random_state=42)
rfq_random.fit(X_train, y_train)
best_rfq = rfq_random.best_estimator_
t_rfq_pred  = best_rfq.predict(X_test, quantiles=0.5)
t_rfq_lower = best_rfq.predict(X_test, quantiles=0.1)
t_rfq_upper = best_rfq.predict(X_test, quantiles=0.9)
t_rfq_coverage = np.mean((y_test.values >= t_rfq_lower) & (y_test.values <= t_rfq_upper)) * 100
t_rfq_r2   = r2_score(y_test, t_rfq_pred)
t_rfq_mae  = mean_absolute_error(y_test, t_rfq_pred)
t_rfq_rmse = np.sqrt(mean_squared_error(y_test, t_rfq_pred))
t_rfq_ape  = (np.abs(y_test.values - t_rfq_pred) / y_test.values * 100).mean()
print(f" Best params : {rfq_random.best_params_}")
print(f" R²: {rfq_r2:.4f} → {t_rfq_r2:.4f}  ({((t_rfq_r2-rfq_r2)/rfq_r2)*100:+.2f}%)")

print("\n All tuning complete.")


# ============================================================
# SECTION 8: TUNED PERFORMANCE VISUALIZATIONS
# ============================================================

# Store tuned full-dataset predictions for trend plots (visualisation only)
df['Best_XGB_Pred']     = best_xgb.predict(X_full)
df['Best_LGBM_Pred']    = best_lgbm.predict(X_full)
df['Best_CatBoost_Pred']= best_cat.predict(X_full)
df['Best_RFQ_Pred']     = best_rfq.predict(X_full, quantiles=0.5)

# --- Tuned R² Leaderboard ---
plt.figure(figsize=(13, 7))
t_labels = ['Baseline\n(Linear)', 'XGBoost\n(Tuned)', 'LightGBM\n(Tuned)',
            'CatBoost\n(Tuned)', 'RF Quantile\n(Tuned)']
t_scores = [base_r2, t_xgb_r2, t_lgbm_r2, t_cat_r2, t_rfq_r2]
t_colors = ['royalblue', 'crimson', 'seagreen', 'darkorange', 'mediumpurple']
t_bars   = plt.bar(t_labels, t_scores, color=t_colors, alpha=0.88, edgecolor='black')
for bar, sc in zip(t_bars, t_scores):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.004,
             f'{sc:.4f}', ha='center', fontweight='bold', fontsize=11)
best_ti = t_scores.index(max(t_scores))
t_bars[best_ti].set_edgecolor('gold')
t_bars[best_ti].set_linewidth(3)
plt.annotate(' Best', xy=(best_ti, t_scores[best_ti] + 0.018),
             ha='center', fontsize=13, color='darkgoldenrod', fontweight='bold')
plt.ylim(0, max(t_scores) + 0.12)
plt.title('Phase 2 - Tuned Algorithm Leaderboard (R²)', fontsize=16)
plt.ylabel('R² Score (Higher is Better)')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()

# --- Tuned Salary Trend by Experience ---
plt.figure(figsize=(14, 7))
t_exp = df.groupby(exp_col).agg({
    'salary_usd':        'mean',
    'Baseline_Pred':     'mean',
    'Best_XGB_Pred':     'mean',
    'Best_LGBM_Pred':    'mean',
    'Best_CatBoost_Pred':'mean',
    'Best_RFQ_Pred':     'mean',
}).sort_index()
plt.plot(t_exp.index, t_exp['salary_usd'],         label='Actual',              color='black',        marker='o', alpha=0.5, lw=2)
plt.plot(t_exp.index, t_exp['Baseline_Pred'],       label='Baseline',            color='royalblue',    linestyle='--')
plt.plot(t_exp.index, t_exp['Best_XGB_Pred'],       label='XGBoost (Tuned)',     color='crimson',      lw=2)
plt.plot(t_exp.index, t_exp['Best_LGBM_Pred'],      label='LightGBM (Tuned)',    color='seagreen',     lw=2)
plt.plot(t_exp.index, t_exp['Best_CatBoost_Pred'],  label='CatBoost (Tuned)',    color='darkorange',   lw=2)
plt.plot(t_exp.index, t_exp['Best_RFQ_Pred'],       label='RF Quantile (Tuned)', color='mediumpurple', lw=2, linestyle='-.')
plt.title('Tuned - Salary Trend by Experience: All Models', fontsize=15)
plt.xlabel('Years of Experience')
plt.ylabel('Average Salary (USD)')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# --- Tuned Feature Importance (XGBoost) ---
t_xgb_imp = pd.Series(best_xgb.feature_importances_, index=X_train.columns).sort_values(ascending=True)
plt.figure(figsize=(10, 6))
t_xgb_imp.plot(kind='barh', color='crimson', edgecolor='black', alpha=0.8)
plt.title('XGBoost Feature Importance (Tuned)', fontsize=13)
plt.xlabel('Relative Importance')
plt.tight_layout()
plt.show()

print("\n XGBoost Tuned - Feature Importance Scores:")
for feat, score in t_xgb_imp.sort_values(ascending=False).items():
    print(f" {feat:<30} : {score:.4f}")


# ============================================================
# SECTION 9: TUNED vs UNTUNED COMPARISON CHARTS
# ============================================================

# --- R² Before vs After ---
fig, axes = plt.subplots(1, 3, figsize=(19, 7), sharey=False)
fig.suptitle('Hyperparameter Tuning Impact - R² Before vs After',
             fontsize=18, fontweight='bold', y=1.02)
for ax, (nm, b_sc, t_sc, dk, lt) in zip(axes, [
    ("XGBoost",     xgb_r2,  t_xgb_r2,  'crimson',     'salmon'),
    ("CatBoost",    cat_r2,  t_cat_r2,  'darkorange',  'peachpuff'),
 ("RF Quantile", rfq_r2,  t_rfq_r2,  'mediumpurple','plum'),
]):
    bars = ax.bar(['Untuned', 'Tuned'], [b_sc, t_sc],
                  color=[lt, dk], edgecolor='black', alpha=0.9, width=0.4)
    for bar, v in zip(bars, [b_sc, t_sc]):
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.0003,
                f'{v:.4f}', ha='center', fontweight='bold', fontsize=12)
    gain = ((t_sc - b_sc) / b_sc) * 100
    ax.set_title(f'{nm}\n({gain:+.2f}% gain)', fontsize=13, fontweight='bold')
    ax.set_ylabel('R² Score')
    ax.set_ylim(min(b_sc, t_sc) - 0.005, max(b_sc, t_sc) + 0.005)
    ax.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()

# --- MAE Before vs After ---
fig, axes = plt.subplots(1, 3, figsize=(19, 7))
fig.suptitle('Tuning Impact - MAE Reduction (Lower is Better)',
             fontsize=16, fontweight='bold', y=1.02)
for ax, (nm, b_m, t_m, dk, lt) in zip(axes, [
    ("XGBoost",     xgb_mae,  t_xgb_mae,  'crimson',     'salmon'),
    ("CatBoost",    cat_mae,  t_cat_mae,  'darkorange',  'peachpuff'),
 ("RF Quantile", rfq_mae,  t_rfq_mae,  'mediumpurple','plum'),
]):
    bars = ax.bar(['Untuned', 'Tuned'], [b_m, t_m],
                  color=[lt, dk], edgecolor='black', alpha=0.9, width=0.4)
    for bar, v in zip(bars, [b_m, t_m]):
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 50,
                f'${v:,.0f}', ha='center', fontweight='bold', fontsize=11)
    delta = b_m - t_m
    ax.set_title(f'{nm}\n(${abs(delta):,.0f} {"saved" if delta >= 0 else "increase"})',
                 fontsize=13, fontweight='bold')
    ax.set_ylabel('MAE (USD)')
    ax.set_ylim(0, max(b_m, t_m) * 1.2)
    ax.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()

# --- LightGBM tuning gain (separate panel) ---
fig, ax = plt.subplots(figsize=(6, 7))
bars = ax.bar(['Untuned', 'Tuned'], [lgbm_r2, t_lgbm_r2],
              color=['lightgreen', 'seagreen'], edgecolor='black', alpha=0.9, width=0.4)
for bar, v in zip(bars, [lgbm_r2, t_lgbm_r2]):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.0003,
            f'{v:.4f}', ha='center', fontweight='bold', fontsize=12)
gain = ((t_lgbm_r2 - lgbm_r2) / lgbm_r2) * 100
ax.set_title(f'LightGBM Tuning\n({gain:+.2f}% gain)', fontsize=13, fontweight='bold')
ax.set_ylabel('R² Score')
ax.set_ylim(min(lgbm_r2, t_lgbm_r2) - 0.005, max(lgbm_r2, t_lgbm_r2) + 0.005)
ax.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()


# ============================================================
# SECTION 10: FINAL SUMMARY
# ============================================================
# All APE values are computed on the test set only.

print("\n" + ""*110)
print(f"{'FINAL RESEARCH PAPER PERFORMANCE SUMMARY':^90}")
print(f"{'Data: ' + source:^90}")
print(""*110)

final_results = [
    # (name,        r2,         mae,        rmse,        ape,        untuned_mae, untuned_rmse)
    ("Baseline",    base_r2,    base_mae,   base_rmse,   base_ape,   base_mae,    base_rmse),
    ("XGBoost",     t_xgb_r2,  t_xgb_mae,  t_xgb_rmse,  t_xgb_ape,  xgb_mae,    xgb_rmse),
    ("LightGBM",    t_lgbm_r2, t_lgbm_mae, t_lgbm_rmse, t_lgbm_ape, lgbm_mae,   lgbm_rmse),
    ("CatBoost",    t_cat_r2,  t_cat_mae,  t_cat_rmse,  t_cat_ape,  cat_mae,    cat_rmse),
    ("RF Quantile", t_rfq_r2,  t_rfq_mae,  t_rfq_rmse,  t_rfq_ape,  rfq_mae,    rfq_rmse),
]

ranked = sorted([r for r in final_results if r[0] != "Baseline"],
                key=lambda x: x[1], reverse=True)
rank_map = {r[0]: i+1 for i, r in enumerate(ranked)}
rank_map["Baseline"] = "-"

print(f"\n{'MODEL':<20}  {'RANK':>4}  {'R²':>6}  {'MAE':>10}  {'RMSE':>10}  {'APE':>7}  {'vs Baseline':>11}")
print(""*90)
for nm, r2, mae, rmse, ape, u_mae, u_rmse in final_results:
    vs   = f"{((r2 - base_r2)/base_r2)*100:+.1f}%" if nm != "Baseline" else "(baseline)"
    rank = rank_map[nm]
    print(f"{nm:<20} {rank!s:>4}  {r2:>6.4f}  "
          f"${mae:>8,.0f} ${rmse:>8,.0f}  {ape:>6.2f}%  {vs:>11}")

print(""*90)

# --- Tuning gains ---
print(f"\n{'TUNING GAINS (Tuned vs Untuned - Test Set)':^90}")
print(""*90)
print(f"{'MODEL':<20}  {'ΔR²':>10}  {'MAE Saved':>12}  {'RMSE Saved':>12}  {'Note'}")
print(""*90)
for nm, u_r2, t_r2, u_mae, t_mae, u_rmse, t_rmse in [
    ("XGBoost",     xgb_r2,  t_xgb_r2,  xgb_mae,  t_xgb_mae,  xgb_rmse,  t_xgb_rmse),
    ("LightGBM",    lgbm_r2, t_lgbm_r2, lgbm_mae, t_lgbm_mae, lgbm_rmse, t_lgbm_rmse),
    ("CatBoost",    cat_r2,  t_cat_r2,  cat_mae,  t_cat_mae,  cat_rmse,  t_cat_rmse),
    ("RF Quantile", rfq_r2,  t_rfq_r2,  rfq_mae,  t_rfq_mae,  rfq_rmse,  t_rfq_rmse),
]:
    dr2   = t_r2  - u_r2
    dmae  = u_mae - t_mae
    drmse = u_rmse - t_rmse
    note  = "Tuning helped" if dr2 > 0 else "Marginal / no gain"
    print(f"{nm:<20} {dr2:>+10.4f}  ${dmae:>10,.0f}  ${drmse:>10,.0f}  {note}")
print(""*90)

# --- Best model callout ---
best_m = max(final_results, key=lambda x: x[1])
print(f"\n Best Model        : {best_m[0]}")
print(f" R²                : {best_m[1]:.4f}")
print(f" Tuned MAE         : ${best_m[2]:,.0f}")
print(f" Tuned RMSE        : ${best_m[3]:,.0f}")
print(f" APE (test set)    : {best_m[4]:.2f}%")
print(f" vs Baseline R²    : {((best_m[1]-base_r2)/base_r2)*100:+.1f}% improvement")

print(f"\n RF Quantile Prediction Interval Coverage (test set):")
print(f" Untuned           : {rfq_coverage:.1f}%    Tuned : {t_rfq_coverage:.1f}%  (nominal target: 80%)")
print(""*110)

# --- EDA stats print (for paper Table I) ---
print("\n" + "="*50)
print("  TABLE I - EDA DESCRIPTIVE STATS (for paper)")
print("="*50)
print(f" Mean       : ${df['salary_usd'].mean():,.0f}")
print(f" Median     : ${df['salary_usd'].median():,.0f}")
print(f" Std Dev    : ${df['salary_usd'].std():,.0f}")
print(f" Min        : ${df['salary_usd'].min():,.0f}")
print(f" Max        : ${df['salary_usd'].max():,.0f}")
print(f" Skewness   : {df['salary_usd'].skew():.4f}")
print(f" Kurtosis   : {df['salary_usd'].kurt():.4f}")
print(f" Q1         : ${df['salary_usd'].quantile(0.25):,.0f}")
print(f" Q3         : ${df['salary_usd'].quantile(0.75):,.0f}")
print(f" IQR        : ${df['salary_usd'].quantile(0.75) - df['salary_usd'].quantile(0.25):,.0f}")
print(f" Total rows : {len(df):,}")
print(f" Train rows : {len(X_train):,}")
print(f" Test rows  : {len(X_test):,}")
print("="*50)
```

```python
# ============================================================
# SECTION 0: IMPORTS & SETUP
# ============================================================
import kagglehub
import numpy as np
import pandas as pd
import os
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from io import StringIO
from matplotlib.patches import Patch

from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
from quantile_forest import RandomForestQuantileRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from scipy.stats import randint as sp_randint

warnings.filterwarnings('ignore')
print(" All libraries loaded.")
```

```python
# ============================================================
# SECTION 1: DATA LOADING (GitHub → Kaggle Fallback)
# ============================================================
# Source 1: GitHub mirror of aijobs.net - weekly updated, CC0,
#           no auth required, reliable raw CSV from Colab
# Source 2: Kaggle dataset - final fallback if GitHub is unreachable

GITHUB_URL = "https://raw.githubusercontent.com/foorilla/ai-jobs-net-salaries/main/salaries.csv"

df_raw = None
source = None

def normalize_aijobs_schema(df):
    """Normalize GitHub/aijobs.net columns into unified pipeline schema."""
    df = df.rename(columns={'salary_in_usd': 'salary_usd'})
    exp_map = {'EN': 1, 'MI': 3, 'SE': 7, 'EX': 12}
    df['years_experience'] = df['experience_level'].map(exp_map).fillna(3)
    df = df.dropna(subset=['salary_usd'])
    df = df[df['salary_usd'] > 0].reset_index(drop=True)
    return df

# --- Source 1: GitHub raw (primary) ---
print(" Fetching latest data from GitHub (foorilla/ai-jobs-net-salaries)...")
try:
    resp = requests.get(GITHUB_URL, timeout=15)
    resp.raise_for_status()
    df_raw = pd.read_csv(StringIO(resp.text))
    missing = [c for c in ['salary_in_usd', 'job_title', 'experience_level']
               if c not in df_raw.columns]
    if missing:
        raise ValueError(f"Unexpected schema - missing columns: {missing}")
    df_raw = normalize_aijobs_schema(df_raw)
    source = " LIVE - GitHub mirror of aijobs.net (weekly updated, CC0)"
    print(f" GitHub data loaded successfully: {df_raw.shape[0]:,} rows")

except Exception as e1:
    print(f" GitHub fetch failed: {e1}")
    print(" Falling back to Kaggle dataset...")

    try:
        path     = kagglehub.dataset_download(
                       'bismasajjad/global-ai-job-market-and-salary-trends-2025')
        csv_path = os.path.join(path, 'ai_job_dataset.csv')
        df_raw   = pd.read_csv(csv_path)
        if 'years_experience' not in df_raw.columns and 'experience_level' in df_raw.columns:
            exp_map = {'EN': 1, 'MI': 3, 'SE': 7, 'EX': 12}
            df_raw['years_experience'] = df_raw['experience_level'].map(exp_map).fillna(3)
        source = " FALLBACK - Kaggle (bismasajjad/global-ai-job-market-and-salary-trends-2025)"
        print(f" Kaggle fallback loaded: {df_raw.shape[0]:,} rows")

    except Exception as e2:
        raise RuntimeError(
            f"Both data sources failed.\n"
            f" GitHub error : {e1}\n"
            f" Kaggle error : {e2}"
        )

df = df_raw.copy()
print(f"\n Data source : {source}")
print(f" Shape        : {df.shape}")
print(f" Salary range : ${df['salary_usd'].min():,.0f} – ${df['salary_usd'].max():,.0f}")
print(f" Columns      : {list(df.columns)}")
```

```python
# ============================================================
# SECTION 2: PREPROCESSING
# ============================================================
# Split data first, then fit encoder on training data only.
# This prevents test-set information from leaking into the encoder.

df_ml = df.copy()
cat_cols = df_ml.select_dtypes(include=['object']).columns.tolist()

# --- Step 1: Split raw (unencoded) data first ---
X_raw = df_ml.drop('salary_usd', axis=1)
y     = df_ml['salary_usd']

X_train_raw, X_test_raw, y_train, y_test = train_test_split(
    X_raw, y, test_size=0.2, random_state=42)

# --- Step 2: Fit encoder on TRAIN only, then apply to both ---
X_train = X_train_raw.copy()
X_test  = X_test_raw.copy()

for col in cat_cols:
    if col not in X_train.columns:
        continue
    le = LabelEncoder()
    X_train[col] = le.fit_transform(X_train[col].astype(str))
    # Map test categories; any unseen label gets encoded as -1
    le_classes = list(le.classes_)
    X_test[col] = X_test[col].astype(str).apply(
        lambda x: le_classes.index(x) if x in le_classes else -1)

# Also encode the full df for EDA visualisations only (not used in evaluation)
X_full = X_raw.copy()
for col in cat_cols:
    if col not in X_full.columns:
        continue
    le_full = LabelEncoder()
    X_full[col] = le_full.fit_transform(X_full[col].astype(str))

print(f" Preprocessing complete | Train: {X_train.shape[0]:,} | Test: {X_test.shape[0]:,}")
print(f" Features: {list(X_train.columns)}")
```

```python
# ============================================================
# SECTION 4: TRAIN ALL MODELS (UNTUNED)
# ============================================================
print("\n" + "="*50)
print("  PHASE 1: UNTUNED MODEL TRAINING")
print("="*50)

exp_col = 'years_experience'

# --- Baseline: Linear Regression ---
print("\n Baseline: Linear Regression...")
baseline = LinearRegression()
baseline.fit(X_train, y_train)
base_pred_test = baseline.predict(X_test)
base_r2   = r2_score(y_test, base_pred_test)
base_mae  = mean_absolute_error(y_test, base_pred_test)
base_rmse = np.sqrt(mean_squared_error(y_test, base_pred_test))
base_ape  = (np.abs(y_test.values - base_pred_test) / y_test.values * 100).mean()
print(f" R²: {base_r2:.4f} | MAE: ${base_mae:,.0f} | RMSE: ${base_rmse:,.0f} | APE: {base_ape:.2f}%")

# --- XGBoost ---
print("\n XGBoost (untuned)...")
xgb = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=5,
                   random_state=42, verbosity=0)
xgb.fit(X_train, y_train)
xgb_pred_test = xgb.predict(X_test)
xgb_r2   = r2_score(y_test, xgb_pred_test)
xgb_mae  = mean_absolute_error(y_test, xgb_pred_test)
xgb_rmse = np.sqrt(mean_squared_error(y_test, xgb_pred_test))
xgb_ape  = (np.abs(y_test.values - xgb_pred_test) / y_test.values * 100).mean()
print(f" R²: {xgb_r2:.4f} | MAE: ${xgb_mae:,.0f} | RMSE: ${xgb_rmse:,.0f} | APE: {xgb_ape:.2f}%")

# --- LightGBM ---
print("\n LightGBM (untuned)...")
lgbm = LGBMRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
lgbm.fit(X_train, y_train)
lgbm_pred_test = lgbm.predict(X_test)
lgbm_r2   = r2_score(y_test, lgbm_pred_test)
lgbm_mae  = mean_absolute_error(y_test, lgbm_pred_test)
lgbm_rmse = np.sqrt(mean_squared_error(y_test, lgbm_pred_test))
lgbm_ape  = (np.abs(y_test.values - lgbm_pred_test) / y_test.values * 100).mean()
print(f" R²: {lgbm_r2:.4f} | MAE: ${lgbm_mae:,.0f} | RMSE: ${lgbm_rmse:,.0f} | APE: {lgbm_ape:.2f}%")

# --- CatBoost ---
print("\n CatBoost (untuned)...")
cat = CatBoostRegressor(iterations=300, learning_rate=0.05,
                        depth=6, random_seed=42, verbose=0)
cat.fit(X_train, y_train)
cat_pred_test = cat.predict(X_test)
cat_r2   = r2_score(y_test, cat_pred_test)
cat_mae  = mean_absolute_error(y_test, cat_pred_test)
cat_rmse = np.sqrt(mean_squared_error(y_test, cat_pred_test))
cat_ape  = (np.abs(y_test.values - cat_pred_test) / y_test.values * 100).mean()
print(f" R²: {cat_r2:.4f} | MAE: ${cat_mae:,.0f} | RMSE: ${cat_rmse:,.0f} | APE: {cat_ape:.2f}%")

# --- RF Quantile Regressor ---
print("\n RF Quantile Regressor (untuned)...")
rfq = RandomForestQuantileRegressor(n_estimators=200, max_depth=15,
                                    min_samples_leaf=5, random_state=42, n_jobs=-1)
rfq.fit(X_train, y_train)
rfq_pred_test = rfq.predict(X_test, quantiles=0.5)
rfq_lower     = rfq.predict(X_test, quantiles=0.1)
rfq_upper     = rfq.predict(X_test, quantiles=0.9)
rfq_coverage  = np.mean((y_test.values >= rfq_lower) & (y_test.values <= rfq_upper)) * 100
rfq_r2   = r2_score(y_test, rfq_pred_test)
rfq_mae  = mean_absolute_error(y_test, rfq_pred_test)
rfq_rmse = np.sqrt(mean_squared_error(y_test, rfq_pred_test))
rfq_ape  = (np.abs(y_test.values - rfq_pred_test) / y_test.values * 100).mean()
print(f" R²: {rfq_r2:.4f} | MAE: ${rfq_mae:,.0f} | RMSE: ${rfq_rmse:,.0f} | APE: {rfq_ape:.2f}%")
print(f" 80% Interval Coverage: {rfq_coverage:.1f}%")

print("\n All untuned models trained.")
```

```python
# ============================================================
# SECTION 7: HYPERPARAMETER TUNING
# ============================================================
print("\n" + "="*50)
print("  PHASE 2: HYPERPARAMETER TUNING")
print("="*50)

# --- Tune LightGBM ---
print("\n Tuning LightGBM...")
lgbm_grid = GridSearchCV(
    LGBMRegressor(random_state=42),
    {'n_estimators':  [100, 200, 300],
     'learning_rate': [0.01, 0.05, 0.1],
     'max_depth':     [3, 5, 10],
     'num_leaves':    [20, 31, 50]},
    cv=3, scoring='r2', verbose=1, n_jobs=-1)
lgbm_grid.fit(X_train, y_train)
best_lgbm = lgbm_grid.best_estimator_
t_lgbm_pred   = best_lgbm.predict(X_test)
t_lgbm_r2   = r2_score(y_test, t_lgbm_pred)
t_lgbm_mae  = mean_absolute_error(y_test, t_lgbm_pred)
t_lgbm_rmse = np.sqrt(mean_squared_error(y_test, t_lgbm_pred))
t_lgbm_ape  = (np.abs(y_test.values - t_lgbm_pred) / y_test.values * 100).mean()
print(f" Best params : {lgbm_grid.best_params_}")
print(f" R²: {lgbm_r2:.4f} → {t_lgbm_r2:.4f}  ({((t_lgbm_r2-lgbm_r2)/lgbm_r2)*100:+.2f}%)")

# --- Tune XGBoost ---
print("\n Tuning XGBoost...")
xgb_grid = GridSearchCV(
    XGBRegressor(random_state=42, verbosity=0),
    {'n_estimators':     [100, 200, 300],
     'learning_rate':    [0.01, 0.05, 0.1],
     'max_depth':        [3, 5, 7],
     'subsample':        [0.7, 0.9, 1.0]},
    cv=3, scoring='r2', verbose=1, n_jobs=-1)
xgb_grid.fit(X_train, y_train)
best_xgb = xgb_grid.best_estimator_
t_xgb_pred   = best_xgb.predict(X_test)
t_xgb_r2   = r2_score(y_test, t_xgb_pred)
t_xgb_mae  = mean_absolute_error(y_test, t_xgb_pred)
t_xgb_rmse = np.sqrt(mean_squared_error(y_test, t_xgb_pred))
t_xgb_ape  = (np.abs(y_test.values - t_xgb_pred) / y_test.values * 100).mean()
print(f" Best params : {xgb_grid.best_params_}")
print(f" R²: {xgb_r2:.4f} → {t_xgb_r2:.4f}  ({((t_xgb_r2-xgb_r2)/xgb_r2)*100:+.2f}%)")

# --- Tune CatBoost ---
print("\n Tuning CatBoost...")
cat_grid = GridSearchCV(
    CatBoostRegressor(random_seed=42, verbose=0),
    {'iterations':    [200, 300, 500],
     'learning_rate': [0.01, 0.05, 0.1],
     'depth':         [4, 6, 8],
     'l2_leaf_reg':   [1, 3, 5]},
    cv=3, scoring='r2', verbose=1, n_jobs=-1)
cat_grid.fit(X_train, y_train)
best_cat = cat_grid.best_estimator_
t_cat_pred   = best_cat.predict(X_test)
t_cat_r2   = r2_score(y_test, t_cat_pred)
t_cat_mae  = mean_absolute_error(y_test, t_cat_pred)
t_cat_rmse = np.sqrt(mean_squared_error(y_test, t_cat_pred))
t_cat_ape  = (np.abs(y_test.values - t_cat_pred) / y_test.values * 100).mean()
print(f" Best params : {cat_grid.best_params_}")
print(f" R²: {cat_r2:.4f} → {t_cat_r2:.4f}  ({((t_cat_r2-cat_r2)/cat_r2)*100:+.2f}%)")

# --- Tune RF Quantile ---
print("\n Tuning RF Quantile (RandomizedSearchCV, n_iter=20)...")
rfq_random = RandomizedSearchCV(
    RandomForestQuantileRegressor(random_state=42, n_jobs=-1),
    {'n_estimators':      sp_randint(100, 400),
     'max_depth':         [10, 15, 20, None],
     'min_samples_leaf': sp_randint(2, 10),
     'max_features':      ['sqrt', 'log2', 0.7, 0.9]},
    n_iter=20, cv=3, scoring='r2', verbose=1, n_jobs=-1, random_state=42)
rfq_random.fit(X_train, y_train)
best_rfq = rfq_random.best_estimator_
t_rfq_pred  = best_rfq.predict(X_test, quantiles=0.5)
t_rfq_lower = best_rfq.predict(X_test, quantiles=0.1)
t_rfq_upper = best_rfq.predict(X_test, quantiles=0.9)
t_rfq_coverage = np.mean((y_test.values >= t_rfq_lower) & (y_test.values <= t_rfq_upper)) * 100
t_rfq_r2   = r2_score(y_test, t_rfq_pred)
t_rfq_mae  = mean_absolute_error(y_test, t_rfq_pred)
t_rfq_rmse = np.sqrt(mean_squared_error(y_test, t_rfq_pred))
t_rfq_ape  = (np.abs(y_test.values - t_rfq_pred) / y_test.values * 100).mean()
print(f" Best params : {rfq_random.best_params_}")
print(f" R²: {rfq_r2:.4f} → {t_rfq_r2:.4f}  ({((t_rfq_r2-rfq_r2)/rfq_r2)*100:+.2f}%)")

print("\n All tuning complete.")
```

```python
# --- Final Performance Summary ---

final_results = [
    ("Baseline (Linear)",   base_r2,   base_mae,   base_rmse,   base_ape,   base_mae,  base_rmse),
    ("XGBoost (Tuned)",     t_xgb_r2,  t_xgb_mae,  t_xgb_rmse,  t_xgb_ape,  xgb_mae,  xgb_rmse),
    ("LightGBM (Tuned)",    t_lgbm_r2, t_lgbm_mae, t_lgbm_rmse, t_lgbm_ape, lgbm_mae, lgbm_rmse),
    ("CatBoost (Tuned)",    t_cat_r2,  t_cat_mae,  t_cat_rmse,  t_cat_ape,  cat_mae,  cat_rmse),
    ("RF Quantile (Tuned)", t_rfq_r2,  t_rfq_mae,  t_rfq_rmse,  t_rfq_ape,  rfq_mae,  rfq_rmse),
]

# Rank models by R2 (baseline excluded from ranking)
ranked   = sorted([r for r in final_results if r[0] != "Baseline (Linear)"],
                  key=lambda x: x[1], reverse=True)
rank_map = {r[0]: i + 1 for i, r in enumerate(ranked)}
rank_map["Baseline (Linear)"] = "-"

W         = 108
sep_heavy = "=" * W
sep_light = "-" * W

print()
print(sep_heavy)
print(f"{'FINAL PERFORMANCE SUMMARY':^{W}}")
print(f"{'Data: ' + source:^{W}}")
print(sep_heavy)

print(
    f"{'MODEL':<25} | {'RANK':>4} | {'R2':>6} | "
    f"{'MAE (Tuned)':>12} | {'RMSE (Tuned)':>12} | "
    f"{'MAE (Untnd)':>11} | {'RMSE (Untnd)':>12} | "
    f"{'APE':>6} | {'vs Baseline':>11}"
)
print(sep_light)

for nm, r2, mae, rmse, ape, u_mae, u_rmse in final_results:
    vs       = f"{((r2 - base_r2) / base_r2) * 100:+.1f}%" if nm != "Baseline (Linear)" else "(baseline)"
    rank     = rank_map[nm]
    mae_str  = f"${mae:>10,.0f}" if nm != "Baseline (Linear)" else f"{'':>12}"
    rmse_str = f"${rmse:>10,.0f}" if nm != "Baseline (Linear)" else f"{'':>12}"
    print(
        f"{nm:<25} | {str(rank):>4} | {r2:>6.4f} | "
        f"{mae_str:>12} | {rmse_str:>12} | "
        f"${u_mae:>9,.0f} | ${u_rmse:>10,.0f} | "
        f"{ape:>5.2f}% | {vs:>11}"
    )

print(sep_heavy)

# --- Tuning gains ---
print()
print(f"{'TUNING GAINS  (Tuned vs Untuned, Test Set)':^{W}}")
print(sep_light)
print(
    f"{'MODEL':<25} | {'Delta R2':>10} | "
    f"{'MAE Saved':>12} | {'RMSE Saved':>12} | {'Conclusion':<20}"
)
print(sep_light)

for nm, u_r2, t_r2, u_mae, t_mae, u_rmse, t_rmse in [
    ("XGBoost",     xgb_r2,  t_xgb_r2,  xgb_mae,  t_xgb_mae,  xgb_rmse,  t_xgb_rmse),
    ("LightGBM",    lgbm_r2, t_lgbm_r2, lgbm_mae, t_lgbm_mae, lgbm_rmse, t_lgbm_rmse),
    ("CatBoost",    cat_r2,  t_cat_r2,  cat_mae,  t_cat_mae,  cat_rmse,  t_cat_rmse),
    ("RF Quantile", rfq_r2,  t_rfq_r2,  rfq_mae,  t_rfq_mae,  rfq_rmse,  t_rfq_rmse),
]:
    d_r2   = t_r2  - u_r2
    d_mae  = u_mae - t_mae
    d_rmse = u_rmse - t_rmse
    note   = "Tuning helped" if d_r2 > 0 else "Marginal / no gain"
    print(
        f"{nm:<25} | {d_r2:>+10.4f} | "
        f"${d_mae:>10,.0f} | ${d_rmse:>10,.0f} | {note:<20}"
    )

print(sep_heavy)

# --- Best model callout ---
best_m = max(final_results, key=lambda x: x[1])
print()
print(f"  Best model     : {best_m[0]}")
print(f"  R2             : {best_m[1]:.4f}")
print(f"  Tuned MAE      : ${best_m[2]:,.0f}")
print(f"  Tuned RMSE     : ${best_m[3]:,.0f}")
print(f"  APE (test set) : {best_m[4]:.2f}%")
print(f"  R2 vs baseline : {((best_m[1] - base_r2) / base_r2) * 100:+.1f}% improvement")
print()
print("  RF Quantile interval coverage (test set, nominal target 80%):")
print(f"    Untuned : {rfq_coverage:.1f}%")
print(f"    Tuned   : {t_rfq_coverage:.1f}%")
print()
print(sep_heavy)
```
